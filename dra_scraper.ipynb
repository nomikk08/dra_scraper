```python
from xml.etree.ElementTree import fromstring, ElementTree
import requests
def public_dra_to_data(entry_num = 1):
    url = f'https://public.dra.gov.pk/rd/registeredDrugsData.svc/prodMains?$filter=((composition%20eq%20null)%20eq%20false)%20and%20((isPublic%20ne%20null)%20and%20(isPublic%20eq%20true))&$orderby=mfgAddress%20desc,composition%20desc&$top={entry_num}'
    response = requests.get(url)
    tree = ElementTree(fromstring(response.content))
    my_root = tree.getroot()
    data = []
    for root in my_root:
        for r in root:
            for i in r.findall('{http://schemas.microsoft.com/ado/2007/08/dataservices/metadata}properties'):

                ids = i.find('{http://schemas.microsoft.com/ado/2007/08/dataservices}id').text
                regNo = i.find('{http://schemas.microsoft.com/ado/2007/08/dataservices}regNo').text
                regNoChar = i.find('{http://schemas.microsoft.com/ado/2007/08/dataservices}regNoChar').text
                brandName = i.find('{http://schemas.microsoft.com/ado/2007/08/dataservices}brandName').text
                composition = i.find('{http://schemas.microsoft.com/ado/2007/08/dataservices}composition').text
                packSize = i.find('{http://schemas.microsoft.com/ado/2007/08/dataservices}packSize').text
                regDate = i.find('{http://schemas.microsoft.com/ado/2007/08/dataservices}regDate').text
                mfgName = i.find('{http://schemas.microsoft.com/ado/2007/08/dataservices}mfgName').text
                mfgAddress = i.find('{http://schemas.microsoft.com/ado/2007/08/dataservices}mfgAddress').text
                mfgCountry = i.find('{http://schemas.microsoft.com/ado/2007/08/dataservices}mfgCountry').text
                importerName = i.find('{http://schemas.microsoft.com/ado/2007/08/dataservices}importerName').text
                importerAddress = i.find('{http://schemas.microsoft.com/ado/2007/08/dataservices}importerAddress').text
                importerCountry = i.find('{http://schemas.microsoft.com/ado/2007/08/dataservices}importerCountry').text
                entityType = i.find('{http://schemas.microsoft.com/ado/2007/08/dataservices}entityType').text


                dic = {
                    'ids': ids,
                    'regNo': regNo,
                    'regNoChar': regNoChar,
                    'brandName': brandName,
                    'Composition': composition,
                    'packSize': packSize,
                    'regDate': regDate,
                    'mfgName': mfgName,
                    'mfgAddress': mfgAddress,
                    'mfgCountry': mfgCountry,
                    'importerName': importerName,
                    'importerAddress': importerAddress,
                    'importerCountry': importerCountry,
                    'entityType': entityType
                }
                data.append(dic)
                
    return data
```

```python
import pandas as pd
d = public_dra_to_data(entry_num = 10)
df = pd.DataFrame(d)
```
